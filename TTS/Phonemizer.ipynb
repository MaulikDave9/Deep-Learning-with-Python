```python
import whisper
import phonemizer
from phonemizer import phonemize
from IPython.display import Markdown, display
import IPython.display as ipd  # For notebook audio playback
import torch
import torchaudio
from pathlib import Path
import numpy as np
import pandas as pd
import sklearn
import nltk
import subprocess
print("All imports work!\n")

print(f"phonemizer: {phonemizer.__version__}")
#print(f"gTTS: {gtts.__version__}")
print(f"whisper: {whisper.__version__}")
print(f"numpy: {np.__version__}")
print(f"pandas: {pd.__version__}")
print(f"scikit-learn: {sklearn.__version__}")
print(f"nltk: {nltk.__version__}")
print(f"torch: {torch.__version__}")
print(f"torchaudio: {torchaudio.__version__}")
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")
```

```python
# Installed from the bash shell
# sudo apt-get install festival festvox-kallpc16k
# sudo apt-get install espeak-ng

# Text to be phonemized
text = """Hello World! I love Deep Learning with PyTorch more when the dependacies are resolved, programs work and models make sense."""

print("\nOriginal text:")
display(Markdown(text))

# Phonemize the text using eSpeak backen
phonemes = phonemize(
    text,
    language='en-us',
    backend='espeak',
    preserve_punctuation=False,
    strip=True
)

print("\nPhonemized text:\n", phonemes)
phonetic_output = phonemize(text, language='en-us', backend='festival', strip=True)

print("\nPhonetic Transcription (Festival Backend):\n", phonetic_output)
```

**eSpeak** - Try to generate speech with eSound.

```python
# Generate speech

output_wav_file = "espeak_output_unionized.wav"
subprocess.run(["espeak", "-v", "en-us", "-w", output_wav_file, text])

print(f"Speech generated and saved as {output_wav_file}")

# Play the audio in the notebook
from IPython.display import Audio, display
display(Audio(output_wav_file))
```
